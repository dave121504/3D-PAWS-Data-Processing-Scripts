```python
import pandas as pd
import xarray as xr
import glob
import os

def convert_dat_to_nc(dat_path, output_dir):
    # Load the .dat file with whitespace delimiter
    df = pd.read_csv(dat_path, delim_whitespace=True)
    
    # Drop unwanted columns if they exist
    columns_to_drop = ['bmp3_pres', 'bmp3_alt', 'bme2_hum', 'vis_light', 'ir_light', 'uv_light']
    df = df.drop(columns=columns_to_drop, errors='ignore')
    
    # Rename columns as per your snippet
    rename_dict = {'mon': 'month', 'min': 'minute'}
    df = df.rename(columns=rename_dict)
    
    # Create a datetime column from year, month, day, hour, minute
    df['time'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute']])
    
    # Set the datetime column as index
    df = df.set_index('time')
    
    # Convert to xarray Dataset
    ds = xr.Dataset.from_dataframe(df)
    
    # Generate output filename based on first timestamp
    date_str = df.index[0].strftime("%Y_%m_%d")
    filename = f"recordings_{date_str}.nc"
    output_path = os.path.join(output_dir, filename)
    
    # Save Dataset to NetCDF
    ds.to_netcdf(output_path)
    print(f"Converted {os.path.basename(dat_path)} → {filename}")

# Define input and output directories
input_dir = "3dpaws dat files/"   # <-- Replace this with your .dat files folder path
output_dir = "converted_nc/"
os.makedirs(output_dir, exist_ok=True)

# Find all .dat files and convert each
dat_files = glob.glob(os.path.join(input_dir, "*.dat"))
for dat_file in dat_files:
    convert_dat_to_nc(dat_file, output_dir)
```

```python


files_paws = sorted(glob.glob('3dpaws_nc/recordings_2025_*.nc'))

for file in files_paws:
    try:
        ds = xr.open_dataset(file)
        time = ds['time'].values

        if len(time) == 0:
            print(f"⚠️  {file} has **no time values**.")
        elif not (time[1:] >= time[:-1]).all():
            print(f"⚠️  {file} has **non-monotonic time values**.")
        else:
            print(f"✅ {file} passed time check.")
    except Exception as e:
        print(f"❌ Failed to open {file}: {e}")
```
